```python
path = pathlib.Path().resolve()
l2_data_path = path.parent.parent / "data" / "l2_data_processed" / "output.csv"
```

```python
data_raw = pd.read_csv(l2_data_path, chunksize = 1000)
```

```python
outdf = pd.DataFrame()

i = 0

for chunk in data_raw:
    df = chunk
    df = df.fillna(0)
    outdf = pd.concat([outdf, df])
    i+=1
    if i >4:
        break
```

```python
unique_addresses = outdf.iloc[:,0].unique().tolist()
```

```python
differences = pd.DataFrame()
count_differences = pd.DataFrame()
```

```python
count = 0
for unique_address in unique_addresses:
    address_rows = outdf.loc[outdf.iloc[:,0]==unique_address, :]
    if len(address_rows)> 1:
        address_rows.loc[:,"count"] = count

        # Get the total number of differences
        tot_diffs = 0
        for comb in itertools.combinations(range(len(address_rows)), 2):
            first_index = comb[0]
            second_index = comb[1]
            diffs = sum(address_rows.iloc[first_index,:] != address_rows.iloc[second_index,:])
            tot_diffs += diffs
        
        address_rows.loc[:,"tot_diffs"] = tot_diffs
            
        differences = pd.concat([differences, address_rows])
    
    count +=1
```
