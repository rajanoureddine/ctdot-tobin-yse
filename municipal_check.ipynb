# Look at old data

```python
import pandas as pd
import pathlib
```

```python
path = pathlib.Path().resolve()
```

```python
old_data = pd.read_csv(path / "ignored-data"/ "vehicles_2022" / "towns01to169_combined.csv",
                       encoding='cp437',
                       chunksize = 1000)
```

```python
# Find norwalk data
i = 0
norwalk_df = pd.DataFrame([])

for chunk in old_data:
    if (i%100 == 0):
        print(f"Now on chunk number {i}")
        
    norwalk_entries = chunk[chunk["recordfrom"].str.contains("Norwalk")]
    num_norwalk = len(norwalk_entries)
    
    if num_norwalk > 0:
        norwalk_df = pd.concat([norwalk_df, norwalk_entries])
        print(f"Now adding {num_norwalk} entries for Norwalk")
    
    i +=1
```
